```
file_path = save_path + "9.txt"
file1 = open(file_path, 'r', encoding="utf8")
Lines = file1.readlines()
```

```
def fix_sentence(line):
    line = line.replace('รายละเอยีดงบประมาณจ าแนกตามงบรายจา่ย', 'รายละเอียดงบประมาณจำแนกตามงบรายจ่าย')
    line = line.replace('รายละเอยีดงบประมาณจ าแนกตามแผนงาน และ ผลผลติ/โครงการ', 'รายละเอียดงบประมาณจำแนกตามแผนงาน และ ผลผลิต/โครงการ')
    line = line.replace('รายละเอยีดงบประมาณจ าแนกตามแผนงาน', 'รายละเอียดงบประมาณจำแนกตามแผนงาน')
    line = line.replace('ท่องเทีย่ว', 'ท่องเที่ยว')
    line = line.replace('กฬีา', 'กีฬา')
    line = line.replace('สงั', 'สัง')
    line = line.replace('มัน่', 'มั่น')
    line = line.replace('ใชจ้า่ย', 'ใช้จ่าย')
    line = line.replace('ใชจ้่าย', 'ใช้จ่าย')
    line = line.replace('งบบคุลากร', 'งบบุคลากร')
    line = line.replace('คลงั', 'คลัง')
    line = line.replace('เงนิ', 'เงิน')
    line = line.replace('ต า่กวา่', 'ต่ำกว่า')
    line = line.replace('ท ัง้สิ้น', 'ทั้งสิ้น')
    line = line.replace('ภณัฑ ์', 'ภัณฑ์')
    line = line.replace('จา่ย', 'จ่าย')
    line = line.replace('ด าเนินงาน', 'ดำเนินงาน')
    line = line.replace('ส าหรบั', 'สำหรับ')
    line = line.replace('ทนุ', 'ทุน')
    line = line.replace('ดา้น', 'ด้าน')
    line = line.replace('รา้ง', 'ร้าง')
    line = line.replace('พฒันา', 'พัฒนา')
    line = line.replace('บคุลากร', 'บุคลากร')
    line = line.replace('ภาครฐั', 'ภาครัฐ')
    line = line.replace('แข่งขนั', 'แข่งขัน')
    line = line.replace('ยทุธศาสตร', 'ยุทธศาสตร์')
    line = line.replace('เ์สรมิ', 'เสริม')
    line = line.replace('ปลดั', 'ปลัด')
    line = line.replace('ู ้', 'ู้')
    line = line.replace('คา่', 'ค่า')
    line = line.replace('จา้ง', 'จ้าง')
    line = line.replace('ทัว่', 'ทั่ว')
    line = line.replace('จ า', 'จำ')
    line = line.replace(' า', 'ำ')
    line = line.replace('บนั', 'บัน')
    line = line.replace('ส านกั', 'สำนัก')
    line = line.replace('องคก์าร', 'องค์การ')
    return line

count = 0
isLock = True
topic = []
price = []
depart = []
subDepart = []
tempDepart = ''
tempSubDepart = ''

for line in Lines:
    line = fix_sentence(line)
    line = line.strip()
    if line.startswith('กระทรวงการคลัง') or line.startswith('กระทรวงการต่างประเทศ') or line.startswith('กระทรวงการท่องเที่ยวและกีฬา') or line.startswith('กระทรวงการพัฒนาสังคมและความมั่นคงของมนุษย์'):
        tempDepart = line
        # print(line)
    if line.startswith('สำนัก') or line.startswith('กรม') or line.startswith('องค์การ') or line.startswith('สถาบัน'):
        tempSubDepart = line
        # print(line)
    if line.startswith('7. รายละเอียดงบประมาณจำแนกตามแผนงาน และ ผลผลิต/โครงการ') and isLock == True:
        count += 1
        print('['+ str(count)+']')
        print('====================================================================================== ' + line + '====================================================================================== ')
        isLock = False
    elif line.startswith('8. รายงานสถานะและแผนการ'):
        print('====================================================================================== ' + line + '====================================================================================== ')
        isLock = True
    elif re.match('\d+\.',line) or re.match('\(\d+\).',line) and isLock == False:
        line = line.split(" ")
        filter_object = filter(lambda x: x != '', line)
        tempLine = list(filter_object)
        if tempLine[len(tempLine)-1] == 'บาท':
            tempTopic = ''
            i = 0
            if len(tempLine) > 3:
                while i < len(tempLine)-2:
                    tempTopic += tempLine[i] + ' '
                    i += 1
                subDepart.append(tempSubDepart)
                depart.append(tempDepart)
                topic.append(tempTopic)
                price.append(int(tempLine[len(tempLine)-2].replace(",","")))
print(topic)
print(price)
```

```
df = pd.DataFrame(list(zip(depart,subDepart,topic, price)), columns =['กระทรวง','สำนัก/กรม','โครงการ','เงิน'])
df
```

```
temp_out_path = out_path+r"/9.csv"
df.to_csv(temp_out_path, index=False, sep=',')
```
